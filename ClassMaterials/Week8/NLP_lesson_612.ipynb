# News article analysis 

## Get text ready for analysis 

```python
textfile=docx2txt.process("Horseracing.docx")
```

```python
textfile
```

```python
#tokenize the text 

tokens=word_tokenize(textfile)
```

```python
tokens
```

```python
# make lower case 
lower_tokens=[tok.lower() for tok in tokens]
```

```python
lower_tokens
```

```python
len(lower_tokens)
```

```python
freq1=nltk.FreqDist(lower_tokens)
```

```python
#create dict of freq - minus any numbers
word_freq1=dict((word,freq) for word, freq in freq1.items() if not word.isdigit())
```

```python
word_freq1
```

```python
sorted1=sorted(word_freq1.items(), key=lambda x:x[1], reverse =True)
```

```python
sorted1
```

## cleaning / wrangling steps 
- pull out the stop words / prepositions 
- pronouns 
- stemming (problem identifying phrases after tokenization)
- remove punctuation

```python
# drop the punctuation , spaces, numeric 
clean1=[word for word in lower_tokens if word.isalpha()]
```

```python
len(clean1)
```

```python
from nltk.corpus import stopwords
```

```python
stop_w=stopwords.words('english')
```

```python
stop_w
```

```python
words_m_stop=[word for word in clean1 if not word in stop_w]
```

```python
len(words_m_stop)
```

```python
freq2=nltk.FreqDist(words_m_stop)
```

```python
word_freq2=dict((word,freq) for word, freq in freq2.items() if not word.isdigit())
```

```python
sorted2=sorted(word_freq2.items(), key=lambda x:x[1], reverse =True)
```

```python
sorted2
```

```python
freq2.most_common(20)
```

## convert clean text into something nltk sees as data 

```python
text3=Text(words_m_stop)
```

```python
type(text3)
```

### collocations - what words are frequently next to eachother 

```python
text3.collocations()
```

### concordance  - fragments containing key words 

```python
text3.concordance("protect")
```

```python
### Lexical dispersion plot - appearance of words / where in text 
text3.dispersion_plot(["horse","racing","racehorses","cruel","death"])
```

```python
df_by_dep=clothingdf.groupby(['Department Name']).agg({'sentiment':'mean'})
```

# Trumps Tweets 

```python
tweets=pd.read_csv('ever_trump.csv')
```

```python
tweets
```

```python
tweets.info()
```

```python
tweets['created_at']=pd.to_datetime(tweets['created_at'])
```

```python
tweets['created_at'].min()
```

```python
tweets['created_at'].max()
```

```python
import spacy 
nlp= spacy.load("en_core_web_sm")
```

```python
#tokenize 
tokens = nlp(''.join(str(tweets.text.tolist())))
```

```python
len(tokens)
```

```python
from collections import Counter
```

```python
items = [x.text for x in tokens.ents]
Counter(items).most_common(20)
```

```python
person_list= [] # empty list 

for ent in tokens.ents:
    if ent.label_ =='PERSON':
        person_list.append(ent.text)
```

```python
person_list
```

```python
person_counts = Counter(person_list).most_common(20)
df_person =pd.DataFrame(person_counts, columns = ['text', 'count'])
#df_person

df_person.plot.barh(x='text', y='count')
```
